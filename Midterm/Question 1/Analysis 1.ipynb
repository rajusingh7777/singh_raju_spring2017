```python
path = '/Users/rajusingh/midterm/data/enron/maildir/' 
```

```python
import os
dirc= os.walk(path)
k=[x[0] for x in dirc]
k
```

```python
def files(path):
    for file in os.listdir(path):
        if not file.startswith('.'):
            if os.path.isfile(os.path.join(path, file)):
                yield file
```

```python
wholelist =[]
for x in k:
    h=files(x)
    j=[x+'/'+y for y in h]
    wholelist.append(j)
#wholelist[:20]
```

```python
flatten = lambda x: [item for sublist in x for item in sublist]
sss=flatten(wholelist)
#sss=sss[:50000]
print(len(sss))
```

```python
from email.parser import Parser 
allPerson =[]
for x in sss:
    with open(x,'r',encoding='utf-8', errors = 'ignore') as f:
        #xx=os.path.join(x,'.txt')
        n=f.read()
        email = Parser().parsestr(n)
        person=email['From'] #or TO
        #If want body uncomment below for each iter:
        #n
        #If want To and From
        if person not in allPerson:
            allPerson.append(person)
allPerson
```
