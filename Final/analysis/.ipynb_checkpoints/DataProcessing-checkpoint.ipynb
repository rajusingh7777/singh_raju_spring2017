```python
df1=pd.DataFrame()
for file in localfileDir:
    if 'ufcx-3fdn' in  file:
        with open(file) as f:
            dfn1 = pd.DataFrame(json.load(f))
            df1 = pd.concat([df1,dfn1])
df1.head()
```

```python
df1.to_csv("/Users/rajusingh/Final/data/ProcessedData/crime_data_2012_to_2015.csv", index=False)
```

```python
df2=pd.DataFrame()
for file in localfileDir:
    if '29yf-ye7n' in  file:
        with open(file) as f:
            dfn2 = pd.DataFrame(json.load(f))
            df2 = pd.concat([df2,dfn2])
df2.head()
```
